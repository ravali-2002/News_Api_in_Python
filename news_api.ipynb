# Languages available

```python
l=LANGUAGES
df=pd.DataFrame()
df["Code"]=l.keys()
df["Language"]=l.values()
lst=[]
for i in range(1,len(df)+1):
    lst.append(i)
df["Num_Code"]=lst
print(tabulate(df,headers=["Code","Language","Num_code"]))
```
